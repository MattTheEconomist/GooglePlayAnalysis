```python
def catVCont(cat, cont):
    thisSer = df.groupby(cat)[cont].mean()
    thisSer = thisSer.sort_values(ascending=False)
    fig = px.bar(thisSer, x=thisSer.index, y=thisSer.values)
    fig.update_layout(
    title="mean"+ cont  +"by" + cat,
    xaxis_title=cat, 
    yaxis_title="mean"+ cont)

    fig.show()
```
