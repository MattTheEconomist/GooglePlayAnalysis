```python
installListOrig = df.Installs.to_list()
installListNew = []

for i in installListOrig:
    i = i.replace("+","").replace(",", "")
    installListNew.append(int(i))

df.Installs = installListNew 
```

```python

sizeList = df.Size.to_list()
sizeLetters= [el[-1] for el in sizeList ]
sizeNumbers=[el[:-1] for el in sizeList]
letterTypes = pd.Series(sizeLetters).value_counts()
letterTypes
```
